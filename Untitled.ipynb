```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

Sales_train = pd.read_csv('Train_UWu5bXk.csv') 

Sales_test = pd.read_csv('Test_u94Q5KV.csv')
```

```python
Sales_test.head()
```

```python
dframes = [Sales_train,Sales_test]
Sales = pd.concat(dframes,ignore_index = True)
print(Sales.shape)
Sales.tail()
```

```python

Sales.describe(include="all")
```

```python
Sales["Qty_Sold"] = (Sales["Item_Outlet_Sales"]/Sales["Item_MRP"])
Sales.head()
```

```python
Sales.isnull().sum()
```

```python
Sales['Item_Weight']=Sales['Item_Weight'].groupby([Sales['Item_Identifier']]).apply(lambda z: z.fillna(z.mean()))
```

```python
Sales.Item_Weight.isnull().sum()
```

```python
import datetime
now = datetime.datetime.now()
now.year

Sales["Outlet_Age"] = now.year - Sales["Outlet_Establishment_Year"]
Sales["Outlet_Age"].head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(Sales.corr())
```

```python
Sales.Outlet_Size.value_counts()
```

```python
Sales.loc[Sales["Outlet_Size"].isnull(),"Outlet_Identifier"].value_counts()
```

```python
Sales['Outlet_Size']=Sales.groupby(['Outlet_Type'])['Outlet_Size'].transform(lambda z: z.fillna(z.mode().ix[0]))

```

```python
Sales.isnull().sum()
```

```python
Sales.describe(include='all')
```

```python
UniqueItems = set(Sales.Item_Identifier)
for each in UniqueItems:
    Sales.loc[(Sales["Item_Identifier"]==str(each)) & (Sales["Item_Visibility"]==0),"Item_Visibility"] = Sales.loc[(Sales["Item_Identifier"]==str(each)) & (Sales["Item_Visibility"]!=0),"Item_Visibility"].mean()
Sales["Item_Visibility"].describe()
```

```python
Sales.head()
```

```python
print(Sales_train.shape)
print(Sales_test.shape)
print(Sales.shape)
```

```python
#replacing values in item type_fat_content
Sales.Item_Fat_Content=Sales.Item_Fat_Content.replace(['Low Fat', 'low fat'], 'LF')
Sales.Item_Fat_Content=Sales.Item_Fat_Content.replace(['Regular'], 'reg')
```

```python
df = Sales.drop(['Outlet_Establishment_Year','Item_Identifier','Outlet_Identifier','Item_Outlet_Sales','Qty_Sold'],axis=1)
```

```python
df.head()
```

```python
X = df.iloc[:].values
```

```python
X_test = X[8523:,:]
```

```python
from sklearn.preprocessing import LabelEncoder
for i in range(len(X[0])):
    le = LabelEncoder()
    le.fit(X[:,i])
    X[:,i] = le.transform(X[:,i])
X=X.astype(int)    
```

```python
y = Sales.iloc[:,3:4].values
```

```python
X = X[:8523,:]
y = y[:8523,:]
```

```python
import xgboost as xgb
```

```python
xg_reg = xgb.XGBRegressor(seed = 123,objective= "reg:linear")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
xg_reg.fit(X_train,y_train)
```

```python
preds = xg_reg.predict(X_validation)
```

```python
from sklearn.metrics import mean_squared_error
# Compute the rmse: rmse
rmse = np.sqrt(mean_squared_error(y_validation, preds))
print("RMSE: %f" % (rmse))
```

```python
y_test_pred = xg_reg.predict(X_test)
len(y_test_pred)
len(X_test)
```
